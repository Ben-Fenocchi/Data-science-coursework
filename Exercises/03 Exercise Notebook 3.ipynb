## 3b) Fit a linear model $f(x) = \beta_0 + \beta_1 x$ to the data

- Print the values of the regression coefficients $\beta_0, \beta_1$ determined using least-squares. 
- Plot the fitted model and the scatterplot on the same plot.
- Compute and print the **MSE** and the $R^2$ coefficient for the fitted model.

All numerical outputs should be displayed to three decimal places.

```python
#Your code here
#polyreg function from exercise 1:
def polyreg(data_matrix, k):
    all_ones = np.ones(np.shape(data_matrix)[0])
    xData = data_matrix[:,0]     
    yData = data_matrix[:,1]      
    n = np.shape(data_matrix)[0] 
    if k>=n:
        k = n -1 
    columns = [all_ones]
    for i in range(k):
         columns.append(xData**(i+1))
    x = np.column_stack(columns)
    betas = np.linalg.lstsq(x, yData, rcond=None)[0]    
    fitLine = x.dot(betas)
    residualValues = yData - fitLine
    return(betas, fitLine, residualValues)
data_matrix = np.column_stack((xvals,yvals))

#calculating fit line
Beta0,Beta1 = polyreg(data_matrix,1)[0]
fitLine = [(value*Beta1 + Beta0) for value in xvals]
beta0 = polyreg(data_matrix,0)[0]
fitLine0 = [(beta0) for value in xvals]

#calculating MSE and R^2
SSE0 = np.linalg.norm(yvals - fitLine0)**2
SSE = np.linalg.norm(yvals - fitLine)**2
MSE = SSE/(np.size(yvals))
R2 = 1 - SSE/SSE0


#plotting
plt.scatter(xvals, yvals,label = "TFL data")
plt.plot(xvals,fitLine,color = "red", label = "Linear fit")
plt.plot(xvals,fitLine0,color = "pink", label = "first order fit")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Tube rider number")
plt.show()

#printing
print('MSE = ', np.round_(MSE, 3), ' \t R^2 = ', np.round_(R2, 3))  
print('Beta0 = ', np.round_(Beta0, 3), ' \t Beta1 = ', np.round_(Beta1, 3))  
```

### 3d) Periodogram

- Compute and plot the peridogram of the residuals. (Recall that the periodogram is the squared-magnitude of the DFT coefficients.)
- Identify the indices/frequencies for which the periogram value exceeds **50%** of the maximum. 


```python
# Your code to compute and plot the periodogram

N = np.size(xvals)
T = 28  # This can be the time interval between any two successive values. (28 or 29 days)

# Note that T is in days

# Compute the squared magnitudes of the DFT coefficients -- this is known as the "periodogram"
pgram = np.abs(np.fft.fft(residuals, N)/N)**2 #We normalize by N, but this is optional.  Periodogram is the squared magnitude of the DFT coefficients
indices = np.linspace(0, (N-1), num = N)
freqs_in_hz = np.fft.fftfreq(N)/T
freqs_in_rads = freqs_in_hz*2*math.pi

plt.subplot(121)
plt.plot(indices, pgram)
plt.xlabel('k')
plt.ylabel('Squared mag. $|G_k|^2/N^2$')
plt.subplot(122)
plt.plot(freqs_in_hz, pgram)
plt.xlabel('frequency (in 1/days)')  # Since units of T is days

plt.tight_layout()
```

```python
# Your code to identify the indices for which the periodogram value exceeds 50% of the maximum
#indices with the strongest sinusoidal components
top_inds = indices[(pgram > 0.5*np.max(pgram))]

print('Top indices:', top_inds)
```

## 3e) To the residuals,  fit a model of the form  

$$ \beta_{1s} \sin(\omega_1 x) + \beta_{1c} \cos(\omega_1 x) + \beta_{2s} \sin(\omega_2 x) + \beta_{2c} \cos(\omega_2 x) + \ldots + \beta_{Ks} \sin(\omega_K x) + \beta_{Kc} \cos(\omega_K x).$$

The frequencies $\omega_1, \ldots, \omega_K$ in the model are those corresponding to the indices identified in Part 2c. (Hint: Each of the sines and cosines will correspond to one column in your X-matrix.)

- Print the values of the regression coefficients obtained using least-squares. 

All numerical outputs should be displayed to three decimal places.

```python
# Your code here

#frequencies of the strongest sinusoidal components
top_freqs_hz = freqs_in_hz[(pgram > 0.5*np.max(pgram))]


w = 2*math.pi*top_freqs_hz  # Freq. in rad/s corresponding to k* is w = 2.pi.k*/(NT)
w=w[:int((np.size(w)/2))] #  frequencies are symmetric, we only care about the first half, see 02 Time series analysis

xRows = []
for omega in w:
    xRows.append(np.sin(omega*xvals))
    xRows.append(np.cos(omega*xvals))

# Now form the X matrix with columns sin(wx) and cos(wx) for x in xvals. First define its transpose
XT = np.vstack(xRows) # np.vstack stacks arrays vertically (same as us working with column stack as we are in the transpose here)
X = np.transpose(XT)

beta_sc = np.linalg.inv(XT.dot(X)).dot(XT).dot(residuals) # Calculating the least-squares coefficients
print('The coefficients beta_s, beta_c are:', beta_sc)
fit_sc = X.dot(beta_sc)



#plotting residuals and fitted sinusoid
plt.subplot(211)
plt.scatter(xvals, residuals, s=20)
plt.plot(xvals, fit_sc, 'r')
plt.title('Scatter plot of de-trended residuals, and fitted sinusoid')
# Zoom in to a few values
plt.subplot(212)
plt.scatter(xvals[100:150], residuals[100:150], s=20)
plt.plot(xvals[100:150], fit_sc[100:150], 'r')

plt.show()
```

### 3f) The combined fit
- Plot the combined fit together with a scatterplot of the data
- Compute and print the final **MSE** and $R^2$ coefficient. Comment on the improvement over the linear fit.

The combined fit, which corresponds to the full model

$$
f(x) = \beta_0 + \beta_1 x + \beta_{s1} \sin(\omega_1 x) + \beta_{c1} \cos(\omega_1 x) + \ldots + \beta_{sk} \sin(\omega_k x) + \beta_{ck} \cos(\omega_k x),
$$

can be obtained by adding the fits in parts 2b) and 2e).

```python
# Your code here
combinedFitLine = fit_sc + fitLine

#plotting
plt.scatter(xvals, yvals,label = "TFL data")
plt.plot(xvals,combinedFitLine,color = "red", label = "Linear + Sinusoidal fit")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Tube rider number")
plt.show()

#calculating MSE and R^2
SSE = np.linalg.norm(yvals - combinedFitLine)**2
MSE = SSE/(np.size(yvals))
R2 = 1 - SSE/SSE0

#Printing
print('MSE = ', np.round_(MSE, 3), ' \t R^2 = ', np.round_(R2, 3))  
```

Mean squared error is reduced by a factor of 45323.636/34486.801 = 1.314. The R^2 value is similar. While this combined model is better it is not much better. Depeniding on what the model is being used for, it could be worth sticking to the linear model.
